## Average Speed and Standard Deviation

The final section of the notebook displays the average speed and st. dev for each speed limit from 20m/s to 32 m/s.  

```python
tronss = Simulation(range(20,33), 480, 100, .1)
```

```python
speeds_list, average_speeds_list = tronss.full_monte()
```

```python
avg_speeds_by_max_speed = []
for i, speed in enumerate(average_speeds_list):
    avg_speed = i + 20, st.mean(speed), st.stdev(speed)
    print(avg_speed)
    avg_speeds_by_max_speed.append(avg_speed)
```
